# Downloading the data from Yahoo Finance

```r
install.packages("quantmod")
```

```r
require('quantmod')
```

```r
# Let's fetch AAPL stock prices
getSymbols("TSLA", src="yahoo", start="2010-01-05", to="2020-01-05")
getSymbols("IBM", src="yahoo", start="2010-01-05", to="2020-01-05")
head(TSLA)
```

```r
head(TSLA$TSLA.Adjusted)
```

```r
# high (Hi), low (lo), closing price(Cl), adjusted closing price(Ad)
plot(Ad(TSLA))
plot(Ad(IBM))
```

```r
mean(Ad(TSLA))
```

```r
# Standard Deviation
sd(Ad(TSLA))
```

```r
# Variance
var(Ad(TSLA))
```

```r
# Square root of the variance is the standard deviation
sqrt(var(Ad(TSLA)))
```

```r
# Correlation between AAPL and IBM
TSLA = TSLA[index(TSLA) > "2015-01-01"]
IBM  = IBM[index(IBM) > "2015-01-01"]
head(TSLA)
head(IBM)
```

```r
# correlation between AAPL and IBM
cor(Ad(IBM), Ad(TSLA))
```

```r
# co-variance between AAPL and IBM
cov(Ad(IBM), Ad(TSLA))
```

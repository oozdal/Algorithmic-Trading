```python
# Measuring the performance of a buy and hold strategy - Max drawdown & calmar ratio

# Import necesary libraries
import yfinance as yf
import numpy as np
import datetime as dt

import matplotlib.pyplot as plt
from matplotlib import pyplot
pyplot.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
# Download historical data for required stocks
ticker = "FB"
SnP = yf.download(ticker,dt.date.today()-dt.timedelta(1825),dt.datetime.today())
```

```python
def CAGR(DF):
    "function to calculate the Cumulative Annual Growth Rate of a trading strategy"
    df = DF.copy()
    df["daily_ret"] = DF["Adj Close"].pct_change()
    df["cum_return"] = (1 + df["daily_ret"]).cumprod()  # cumprod -> cumulative product
    n = len(df)/252   # 252 is the total number of trading days per year.
    CAGR = (df["cum_return"][-1])**(1/n) - 1  # 1/n refers to 1/years
    return CAGR
```

```python
def max_dd(DF):
    "function to calculate max drawdown"
    df = DF.copy()
    df["daily_ret"] = DF["Adj Close"].pct_change()
    df["cum_return"] = (1 + df["daily_ret"]).cumprod()
    df["cum_roll_max"] = df["cum_return"].cummax()
    df["drawdown"] = df["cum_roll_max"] - df["cum_return"]
    df["drawdown_pct"] = df["drawdown"]/df["cum_roll_max"]
    max_dd = df["drawdown_pct"].max()
    return max_dd
```

```python
def calmar(DF):
    "function to calculate calmar ratio"
    df = DF.copy()
    clmr = CAGR(df)/max_dd(df)
    return clmr
```

```python
DF = SnP.copy()
DF["daily_ret"] = DF["Adj Close"].pct_change()
DF["cum_return"] = (1 + DF["daily_ret"]).cumprod()
DF["cum_roll_max"] = DF["cum_return"].cummax()
DF["cum_return"].plot()
```

```python
print("max drawdown:", max_dd(DF))
```

```python
print("calmar ratio:", calmar(DF))
```

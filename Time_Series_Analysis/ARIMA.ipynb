# ARIMA model example

```r
# simulate an ARIMA process with AR(1), I(1) and MA(1) and the coefficients
x = arima.sim(n=2000, model=list(c(1,1,1), ar=0.4, ma=0.8))
```

```r
plot(x)
```

```r
# We would like to approximate the process ARIMA
model =  arima(x, order=c(1,1,1))
model
```

```r
model$coef
```

```r
acf(resid(model))
```

```r
# Ljung-Box test
Box.test(resid(model),lag=25, type="Ljung-Box")
```

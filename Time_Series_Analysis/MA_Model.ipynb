```r
require(quantmod)
```

```r
# we download AAPL stock prices from Yahoo Finance
getSymbols("AAPL", src="yahoo")
```

```r
# we weant to plot the adjusted closing prices
plot(Ad(AAPL))
```

```r
# log daily returns of AAPL stock (for adjusted closing price)
aaplreturn = diff(log(Ad(AAPL)))
```

```r
plot(aaplreturn)
acf(aaplreturn, na.action=na.omit)
```

```r
# we set AR=0 and integrated component=0 so it is a MA(3)
applreturn.ma <- arima(aaplreturn, order=c(0,0,3))
applreturn.ma
```

```r
# residual values: the difference between the actual values x(t)
# and the values predicted by a given model x'(t)
# y(t) = x(t) - x'(t)  residual error series
head(applreturn.ma$res)
```

```r
# drop the first value of the residual value because it is NA
head(applreturn.ma$res[-1])
```

```r
# y = x - x' the residual series: if the model is good there is o serial correlation in y(t)
acf(applreturn.ma$res[-1])
```

```r
require(lmtest)
```

```r
# We can construct an ARMA(p,q) simulation with the AR and MA components
x= arima.sim(n=2000, model=list(ar=0.5, ma=0.1))  # ar = alpha, ma = beta parameter
head(x)
```

```r
plot(x)
```

```r
# fit an ARIMA model with AR(1) and MA(1)
model = arima(x, order=c(1,0,1))   # order of autoregressive part=1, integerated part=0, moving average order=1
model
```

```r
coeftest(model)
```

```python
import datetime as dt
import yfinance as yf
import pandas as pd
```

```python
stocks = ["AMZN", "MSFT", "FB", "GOOG"]
start  = dt.datetime.today()-dt.timedelta(3650)
end    = dt.datetime.today()
c1_price = pd.DataFrame() # empty DataFrame
oh1cv_data = {} # empty dictionary
```

```python
# Looping over tickers and creating a dataframe with close prices
for ticker in stocks:
    c1_price[ticker] =  yf.download(ticker, start, end)["Adj Close"]
```

```python
c1_price
```

```python
c1_price.fillna(0) # Filling NaN values
```

```python
c1_price.fillna({"FB":0, "GOOG":1}) # Filling NaN values
```

```python
c1_price.fillna(method='bfill') # Filling NaN values: bfill refers to backfill
```

```python
c1_price.fillna(method='bfill', axis=1) # Filling NaN values: axis=1 takes the values of the following column
```

```python
c1_price.fillna(method='bfill', axis=1, inplace=True) # inplace=True fills the actual dataframe.
c1_price
```

```python
# Looping over tickers and creating a dataframe with close prices
for ticker in stocks:
    c1_price[ticker] =  yf.download(ticker, start, end)["Adj Close"]
```

# Modeling assets with autoregressive model

```r
install.packages("quantmod")
```

```r
require("quantmod")
```

```r
# we download IBM stock prices from Yahoo Finance
getSymbols("IBM", src="yahoo")
```

```r
plot(Ad(IBM))
```

```r
# let's calculate the log daily return
returns = diff(log(Ad(IBM)))
```

```r
plot(returns)
```

```r
acf(returns, na.action=na.omit)
```

```r
# Let's build autoregressive model
model = ar(returns, na.action=na.omit)
model
```

```r
model$order
```

```r
model$ar # alpha coefficients
```

```r
model$asy.var # variances
```

```r
# residual series = actual values - predicted by the model
acf(model$resid, na.action =  na.omit)
```

```r
# Let's do the same analysis for ^GSPC
# we download IBM stock prices from Yahoo Finance
getSymbols("^GSPC", src="yahoo")
```

```r
plot(Ad(GSPC))
```

```r
# let's calculate the log daily return
returns = diff(log(Ad(GSPC)))
```

```r
plot(returns)
```

```r
acf(returns, na.action=na.omit)
```

```r
# Let's build autoregressive model
model = ar(returns, na.action=na.omit)
model
```

```r
model$order # Model with 18 parameters! Overfitting!!
```

```r
# residual series = actual values - predicted by the model
acf(model$resid, na.action =  na.omit)
```

```r
# Conclusion: Autoregressive model works well when dealing with IBM
# but it is not working well when we want to understand and grasp S&P 500
# related index
```
